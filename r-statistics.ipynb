```r
Data_Cars <- mtcars

median(Data_Cars$wt)
```

# Mode

The mode value is the value that appears the most number of times.

R does not have a function to calculate the mode. However, we can create our own function to find it.

If we take a look at the values of the wt variable (from the mtcars data set), we will see that the numbers 3.440 are often shown:



```r
Data_Cars <- mtcars

names(sort(-table(Data_Cars$wt)))[1]
```

This code computes the most frequently occurring value in the 'wt' column of the mtcars dataset and retrieves its name.

Let's break down the code:

1. `Data_Cars <- mtcars`: This line assigns the mtcars dataset to the variable `Data_Cars`.

2. `table(Data_Cars$wt)`: This creates a frequency table of the 'wt' column in the mtcars dataset, counting how many times each unique value appears.

3. `sort(-table(Data_Cars$wt))`: This sorts the frequency table in descending order of counts. The `-` sign indicates reverse sorting.

4. `names(...)`: This retrieves the names of the sorted table, which corresponds to the unique values of the 'wt' column.

5. `[1]`: This index selects the first (i.e., most frequent) value from the sorted table.

So, the entire expression is finding the most frequent value in the 'wt' column of the mtcars dataset.

```r
Data_Cars <- mtcars

mode(Data_Cars$wt)
```

# R Percentiles

# Percentiles

Percentiles are used in statistics to give you a number that describes the value that a given percent of the values are lower than.

If we take a look at the values of the wt (weight) variable from the mtcars data set:

Observation of wt (weight)

```r
Data_Cars <- mtcars

# c() specifies which percentile you want
quantile(Data_Cars$wt, c(0.75))
```

```r
Data_Cars <- mtcars

quantile(Data_Cars$wt)
```

# Quartiles
Quartiles are data divided into four parts, when sorted in an ascending order:

The value of the first quartile cuts off the first 25% of the data
The value of the second quartile cuts off the first 50% of the data
The value of the third quartile cuts off the first 75% of the data
The value of the fourth quartile cuts off the 100% of the data
Use the quantile() function to get the quartiles.
