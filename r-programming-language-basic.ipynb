# R Nested Functions

There are two ways to create a nested function:

Call a function within another function.
Write a function within a function.

```r
#Call a function within another function
my_function <- function(x,y){
    a <- x * y
    return(a)
}
my_function(my_function(2,4), my_function(3,5))
```

```r
#Write a function within a function

Outer_func <- function(x) {
  Inner_func <- function(y) {
    a <- x + y
    return(a)
  }
  return (Inner_func)
}
output <- Outer_func(3) # To call the Outer_func
output(5)
```

# R Function Recursion

```r
tri_recursion <- function(k){
    if(k > 0){
        result <- k + tri_recursion(k -1)
        print(result)
    }
    else{
        result = 0
        return(result)
    }
}

tri_recursion(10)
tri_recursion(0)
tri_recursion(1)
tri_recursion(2)

```

```r
#fund factorial through recusion
factorial <- function(n) {
    if (n == 0) {
        return(1)  # Base case: factorial of 0 is 1
    } else {
        return(n * factorial(n - 1))  # Recursive case: factorial of n is n times factorial of (n-1)
    }
}
result <- factorial(5)
print(result)  # Output will be 120
```

# The Global Assignment Operator

```r
#If you use the assignment operator <<-, the variable belongs to the global scope

my_function <- function() {
txt <<- "fantastic"
  paste("R is", txt)
}

my_function()

print(txt)
```

```r
txt <- "awesome"
my_function <- function() {
  txt <<- "fantastic"
  paste("R is", txt)
}

my_function()

paste("R is", txt)
```
